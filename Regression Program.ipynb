### Importing Required Pakages
    
</div>

```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import linear_model
%matplotlib inline
```

<div class="alert alert-block alert-info">

### Creating a Dataframe using the REGRESSION.csv
    
</div>

```python
df = pd.read_csv("REGRESSION.csv")
df
```

<div class="alert alert-block alert-info">

### Plottng the Scatter Plot
    
</div>

```python
plt.scatter(df.Study,df.Marks,color='blue')
```

<div class="alert alert-block alert-info">

### Importing train_test_split for difurcating the data for training and testing
    
</div>

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['Study']],df.Marks,train_size=0.8)
```

<div class="alert alert-block alert-info">

### Checking the Train & Test Data
    
</div>

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

<div class="alert alert-block alert-info">

### Importing Linear Regression and Creating a Linear Regression Object
    
</div>

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

<div class="alert alert-block alert-info">
    
### Fitting the Training Data into reg (Linear Rgegression Object)
    
</div>

```python
reg.fit(X_train, y_train)
```

<div class="alert alert-block alert-success">

## ***Predicting Marks with Linear Regression***
    
</div>

```python
predictions = reg.predict([[10]])
print(*predictions,"%")
```

***

<div class="alert alert-block alert-info">

### Coefficient
    
</div>

```python
coefficient = reg.coef_
print(*coefficient)
```

<div class="alert alert-block alert-info">

### Y - Intercept
    
</div>

```python
intercept = reg.intercept_
print(intercept)
```

<div class="alert alert-block alert-success">

## Y = m*x + b
    
</div>

```python
y = 157 * reg.coef_ + reg.intercept_
print(*y,"%")
```
